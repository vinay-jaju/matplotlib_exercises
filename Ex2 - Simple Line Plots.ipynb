```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
fig = plt.figure()
ax = plt.axes()
```

```python

x=np.linspace(0,10,100)
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
```

```python
plt.plot(x, np.sin(x - 0), color='blue')        # specify color by name
plt.plot(x, np.sin(x - 1), color='g')           # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')        # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 to 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # all HTML color names supported
```

```python
plt.plot(x,x+0,linestyle='solid')
plt.plot(x,x+1,linestyle='dashed')
plt.plot(x,x+2,linestyle='dashdot')
plt.plot(x,x+3,linestyle='dotted')

# for short
plt.plot(x,x+4,linestyle='-')
plt.plot(x,x+5,linestyle='--')
plt.plot(x,x+6,linestyle='-.')
plt.plot(x,x+7,linestyle=':');

```

```python
#Being Lazy
plt.plot(x,x,'-g')
plt.plot(x,x+1,'--c')
plt.plot(x,x+2,'-.k')
plt.plot(x,x+3,':r');
```

```python
plt.plot(x,np.sin(x))
plt.xlim(-1,11)
plt.ylim(-1.5,1.5)
```

```python
plt.plot(x, np.sin(x))
plt.axis([-1, 11, -1.5, 1.5]);
```

```python
plt.plot(x, np.sin(x))
plt.axis('tight');
```

```python
plt.plot(x, np.sin(x))
plt.axis('equal');
```

```python
plt.plot(x, np.sin(x))
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)");
```

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')

plt.legend()
```

## Aside: Matplotlib Gotchas:

While most plt functions translate directly to ax methods (such as plt.plot()  ax.plot(), plt.legend()  ax.legend(), etc.), this is not the case for all commands. In particular, functions to set limits, labels, and titles are slightly modified. For transitioning between MATLAB-style functions and object-oriented methods, make the following changes:

* plt.xlabel()  ax.set_xlabel()
* plt.ylabel()  ax.set_ylabel()
* plt.xlim()  ax.set_xlim()
* plt.ylim()  ax.set_ylim()
* plt.title()  ax.set_title()

```python
#Using set method
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim=(0, 10), ylim=(-2, 2),
       xlabel='x', ylabel='sin(x)',
       title='A Simple Plot');
```
